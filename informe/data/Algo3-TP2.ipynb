```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import random
import glob
import os
```

# Problema 1

```python
df = pd.read_csv('delivery.csv')
```

```python
dfA = df[(df.Ciudades > 0) & (df.Rutas == 5) & (df.TotalPremium == 3) & (df.LimitePremium == 2)]

sns.lmplot(x='Ciudades', y='Nanosegundos', data=dfA, size=7, aspect=1.4, order=2)
plt.show()
```

```python
dfB = df[(df.Ciudades == 10) & (df.Rutas > 5) & (df.TotalPremium == 5) & (df.LimitePremium == 2)]

sns.lmplot(x='Rutas', y='Nanosegundos', data=dfB, size=7, aspect=1.4)
plt.show()
```

```python
dfC = df[(df.Ciudades == 10) & (df.Rutas == 20) & (df.TotalPremium > 1) & (df.LimitePremium == 2)]

g = sns.lmplot(x='TotalPremium', y='Nanosegundos', data=dfC, size=7, aspect=1.4)
g.set(ylim=(0,17000))
plt.show()
```

```python
dfD = df[(df.Ciudades == 10) & (df.Rutas == 30) & (df.TotalPremium > 0) & (df.LimitePremium > 0)]

sns.lmplot(x='LimitePremium', y='Nanosegundos', data=dfD, size=7, aspect=1.4, order=2)
plt.show()
```
